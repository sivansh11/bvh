```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
def extract_data(path):
    with open(path, 'r') as file:
        data_start = False
        all_data = []
        for line in file.readlines():
            if "#starting test" in line:
                data_start = True
                continue
            if "average render time:" in line:
                continue
            if data_start:
                data = float(line.replace("render took: ", ""))
                all_data.append(data)
        return all_data
```

```python
configs = {}
configs["config_1"] = extract_data("../bvh/build/config_1.txt")
configs["config_2"] = extract_data("../bvh/build/config_2.txt")
configs["config_3"] = extract_data("../bvh/build/config_3.txt")
configs["config_4"] = extract_data("../bvh/build/config_4.txt")
configs["config_5"] = extract_data("../bvh/build/config_5.txt")
configs["config_6"] = extract_data("../bvh/build/config_6.txt")
configs["config_7"] = extract_data("../bvh/build/config_7.txt")
configs["config_8"] = extract_data("../bvh/build/config_8.txt")
configs["config_9"] = extract_data("../bvh/build/config_9.txt")
configs["config_10"] = extract_data("../bvh/build/config_10.txt")
configs["config_11"] = extract_data("../bvh/build/config_11.txt")
configs["config_12"] = extract_data("../bvh/build/config_12.txt")
configs["config_13"] = extract_data("../bvh/build/config_13.txt")
configs["config_14"] = extract_data("../bvh/build/config_14.txt")
configs["config_15"] = extract_data("../bvh/build/config_15.txt")
configs["config_16"] = extract_data("../bvh/build/config_16.txt")
```

```python
def plot(data, name):
    plt.figure()
    x = np.arange(len(data))
    plt.plot(x, data)
    plt.xlabel("iterations")
    plt.ylabel("time took")
    plt.title(name)
```

```python
def remove_outliers_zscore(df, column_name, threshold=3):
    mean = df[column_name].mean()
    std = df[column_name].std()
    upper_limit = mean + threshold * std
    lower_limit = mean - threshold * std
    return df[(df[column_name] <= upper_limit) & (df[column_name] >= lower_limit)]

def remove_outliers_iqr(df, column_name):
  q1 = df[column_name].quantile(0.25)
  q3 = df[column_name].quantile(0.75)
  iqr = q3 - q1
  upper_limit = q3 + 1.5 * iqr
  lower_limit = q1 - 1.5 * iqr
  return df[(df[column_name] <= upper_limit) & (df[column_name] >= lower_limit)]

```
